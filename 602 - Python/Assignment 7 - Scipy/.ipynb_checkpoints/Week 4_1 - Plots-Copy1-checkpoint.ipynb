```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
river = pd.read_csv("https://raw.githubusercontent.com/charleyferrari/CUNY_DATA608/master/lecture4/Data/riverkeeper_data_2013.csv", sep=",")
river['Date'] = pd.to_datetime(river.Date)

river.set_index('Date', inplace= True)

river['EnteroCount'] = pd.to_numeric(pd.Series(river['EnteroCount'].str.replace('>|<',"")))

#print(river['EnteroCount'] )

river[river.index < '20080101']

#river.loc['20100102':'20130104']
river.describe()
river.head()
```

```python
#Best places to swim

river[river['SampleCount'] > 110]
river['DayRain_ml'] = (river['FourDayRainTotal']/4)*16.3872
river['average'] = river['EnteroCount']/river['DayRain_ml']
river[river['average'] > 0.3]
river.describe()
print(river[river['Site']=='Hudson above Mohawk River'].describe())
```

```python
#print(river[river['Site']=='Hudson above Mohawk River'].describe())
#river["Site"] = river["Site"].astype('category')
#print(river.groupby('Site').describe())
river_new = river.groupby('Site').agg(['mean','count'])
(river_new['EnteroCount']['count']< 5) & (river_new['EnteroCount']['mean'] > 110)

#np.where(river_new['EnteroCount']['count'] < 5 )

river_new['swim'] = np.where(
    ((river_new['EnteroCount']['count']< 5) & (river_new['EnteroCount']['mean'] > 110)) |
    ((river_new['EnteroCount']['count'] > 5) & (river_new['EnteroCount']['mean'] > 30)), 
    'unacceptable', 'acceptable')
river_new
```
